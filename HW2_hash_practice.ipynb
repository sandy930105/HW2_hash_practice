```python
def analyze_with_hash(filename):
    with open(filename, 'r') as file:
        words = [line.strip() for line in file if line.strip()]

    word_dict = {}  # 這就是 hash table（dict）

    for word in words:
        if word in word_dict:
            word_dict[word] += 1
        else:
            word_dict[word] = 1

    # 顯示總數
    print(f"\n🔢 不重複英文字總數：{len(word_dict)}")

    # 顯示每個英文字出現次數
    print("\n📋 每個英文字的出現次數：")
    for word, count in word_dict.items():
        print(f"{word}: {count}")

    # 排序並畫圖
    sorted_items = sorted(word_dict.items(), key=lambda x: x[1], reverse=True)
    labels, values = zip(*sorted_items)

    plt.figure(figsize=(10, 6))
    plt.bar(labels, values, color='lightcoral')
    plt.title("英文字出現次數直方圖（使用 dict/hash）")
    plt.xlabel("英文字")
    plt.ylabel("出現次數")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.show()

# 用 list(uploaded.keys())[0] 取得上傳檔案
analyze_with_hash(list(uploaded.keys())[0])
```
